```python
import pandas as pd
import numpy as np
```

```python
chunks = []
for chunk in pd.read_csv('data/transactions.csv.gz', compression='gzip', chunksize=2*(10**7), engine="c"):
    #pd.concat([transactionsdf_raw, chunk], ignore_index=True)
    df = chunk[chunk.company==103700030]
    chunks.append(df)
```

```python
transactionsdf_raw = pd.concat(chunks, ignore_index=True)
```
